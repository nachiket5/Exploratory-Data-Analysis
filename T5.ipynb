## Visualization

```python
# Matches per season
plt.subplots(figsize=(20,10))

sns.countplot(x = 'season', data = matches, palette = 'deep')

plt.title('Matches per Season',fontsize=35)
plt.xlabel('Season',fontsize=15)
plt.ylabel('Matches',fontsize=15)

plt.show()
```

```python
# Matches per team
num_matches = pd.concat([matches['team1'], matches['team2']])
# print(num_matches)
num_matches = num_matches.value_counts()
print(num_matches)
plt.figure(figsize=(20,10))

plt.bar(x=num_matches.index,height=num_matches.values,color='Blue')
# sns.countplot(x=num_matches.index,y= num_matches.values,data=matches,palette='rocket_r')

plt.title('Matches per Team',fontsize=35)
plt.xlabel('Team',fontsize=15)
plt.ylabel('Matches',fontsize=15)
plt.xticks(rotation=90,fontsize=15)
    
plt.show()
```

```python
# Visualization of how many matches were played in which venue.
plt.subplots(figsize=(20,10))

sns.countplot(x ='venue',data=matches,palette='deep',order=matches['venue'].value_counts().index)

plt.title('Indian Premier League Venue',fontsize=35)
plt.xlabel('Stadium',fontsize=15)
plt.ylabel('Matches',fontsize=15)
plt.xticks(rotation=270,fontsize=15)

plt.show()
```

```python
seasonw = matches.drop_duplicates('season', keep='last')
seasonw = seasonw[['season', 'winner']]
seasonw.sort_values('season',inplace=True)
seasonw.reset_index(inplace=True, drop=True)
```

```python
print(seasonw)
```

```python
# Championships per Team
plt.subplots(figsize=(20,10))

sns.countplot('winner', data = seasonw, palette = 'rocket_r')

plt.title('Championship',fontsize=30)
plt.xlabel('Teams',fontsize=15)
plt.ylabel('Season won',fontsize=15)

plt.show()
```

```python
top_players = matches.player_of_match.value_counts()[:10]
```

```python
# Top 10 players
plt.figure(figsize=(20,10))

sns.barplot(x=top_players.index,y=top_players, palette="colorblind")

plt.title('Top players of the match Winners',fontsize=35)
plt.xlabel('Player',fontsize=20)
plt.ylabel('Match',fontsize=20)

plt.show()
```
